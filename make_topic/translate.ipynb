```python
from deep_translator import GoogleTranslator
import nltk
import pandas as pd
from nltk.tokenize import word_tokenize
from tqdm import tqdm

# NLTK 패키지 다운로드 (최초 1회만 필요)
nltk.download('punkt')
```

```python
df = pd.read_csv('dewey.csv')
translator = GoogleTranslator(source='en', target='ko')

df.fillna('', inplace=True)
df
```

```python
#번역
#소요 시간: 75m

# 번역함수 생성
def translate_text(text):
    if text:
        return translator.translate(text)
    return text

translated_lists = [] #임시 저장

for index, row in tqdm(df.iterrows(), desc='df ~3246'):
    translated_row = [
        translate_text(row['주류']),
        translate_text(row['강목']),
        translate_text(row['요목명']),
        translate_text(row['세목명'])
    ]
    translated_lists.append({'ddc' : row['세목번호'],  
                            '주류(한글)': translated_row[0],
                            '강목(한글)': translated_row[1],
                            '요목(한글)': translated_row[2],
                            '세목(한글)': translated_row[3]})
    
```

```python
df = pd.DataFrame(translated_lists)
df
```

```python
df.to_csv('translated.csv')
```
